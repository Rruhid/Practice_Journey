```python
# 1)
import pandas as pd  

data = pd.read_csv("data_practice/Survey_Data.csv")
data 
```

```python
data['collection_date'].unique() 
```

```python
#Show total number of unique values in a column 
#data['collection_date'].nunique() 

#2) Total number of unique values with their count on each day 
data['collection_date'].value_counts(ascending = False)
```

```python
# 3) What proportion of the total respondents were aged less than 45?  
data['age'].unique()
data['age'].value_counts()

#data[data['age']>45] # gives error
data.age.dtypes 
# Here we have '24ko' - type is object in order to work with number data we should change it and whole dtype 

data.age.replace({'24ko': 24}, inplace=True)
data.age=data.age.astype(int)
data.age.dtype
data[data['age']<45]
perc = (6399/6867) * 100
round(perc)
```

```python
# 4) Create a new column in the dataframe "age_group". This column should contain the age group the respondent belongs to. The age groups are 18-25, 25-45, and 45+.

#  make a copy if needed
data.head(2)
copy_file = data.copy()
copy_file.head()

#Adding New Column 
#data.insert(10, 'age_group', data.age) 
data['age_group'] = data['age_group'].astype(int)
#Filtering
data1=data[(data.age_group >= 18) & (data.age_group <25)]
data1.age_group.unique()
data1.age_group = "18-25"
data1.head(2)



data2= data[(data.age_group >=25) & (data.age_group <45)]
data2.head(2)
data2.age_group.unique()
data2.age_group = "25-45"
data2.head(2)


data3 = data[data.age_group >= 45]
data3.age_group.unique()
data3.age_group = "45+"
data3.head(2)


data_conclusion = pd.concat([data1, data2, data3])
data_conclusion

data_conclusion.age_group.unique()
 
```

```python
# 5) How many samples were collected for each age-group? Which age-group had the most samples? 
data_conclusion['age_group'].value_counts()
```

```python
# 6) What proportion of the respondents had opted for the RJD party in both the Vote_Now and the Past_Vote questions.

data_conclusion.columns
data_conclusion.Vote_Now.unique()
data_conclusion.Past_Vote.unique()

data_conclusion[(data_conclusion['Vote_Now'] == 'RJD' ) & (data_conclusion['Past_Vote'] == 'RJD')]
perc = (811 / 6867 ) * 100 
round(perc)
```

```python
data_conclusion['CM_satisfaction'].unique()
```

```python
# 7) For each day of sample collection, determine the proportion of respondents who were fully satisfied with the performance of the CM. So if there were a total of 1000 samples on day 1 and 300 out of those said they were fully satisfied, then our answer for that day would be 0.3.
data_conclusion.columns
cm_data = data_conclusion[data_conclusion['CM_satisfaction'] == 'Fully Satisfied']

a = cm_data.collection_date.value_counts()
a
b= data['collection_date'].value_counts(ascending = False)
b
c= (a/b) * 100
c
```

```python
# 8) In a similar fashion create a day-wise proportion of respondents that opted fully dissatisfied with their MLA. Create a line plot of the result with date on x-axis and proportions on the y-axis.

 
data.head()
pla_data = data[data['MLA_satisfaction'] == 'Fully Dissatisfied']
d = pla_data.collection_date.value_counts()
d

e=(d/b)*100
round(e) 
g = pd.DataFrame(e)
#g

g.collection_date.plot(kind='line', figsize=(20,10), color='red')
```

```python
# 9) Create a pivot-table (or crosstab) with index as Past_Vote, Column as Vote_Now and cell values as the count of samples.


data[['Past_Vote','Vote_Now']]
data.pivot_table(index='Past_Vote', columns='Vote_Now', aggfunc = 'count')
```

```python
# 10) Repeat the above question with the cell values as the sum of “weight”.
data.head(2)
data.pivot_table(index='Past_Vote', columns='Vote_Now', values = 'weight', aggfunc='sum')
```

```python
# 11) Create a dataframe by performing a group by over age_group and calculate the count of total samples under each age_group.

data_conclusion.head(2)
df1 = data_conclusion.groupby('age_group').count()
df1
```

```python
# 12) Create a dataframe by performing a group by over age_group and finding the count of total samples for each age_group that opted for the JD(U) party in Vote_Now.
data_conclusion.head()
data_conclusion.Vote_Now.unique()
data_jdu = data_conclusion[data_conclusion.Vote_Now == 'JD(U)']
#data_jdu
df2 = data_jdu.groupby('age_group').count()
df2
```

```python
# 13) Join/Merge the two dataframes from questions 12 and 13 with the common column as age_group. 

mergge = pd.merge(df1, df2, on = 'age_group')
mergge.columns
```
